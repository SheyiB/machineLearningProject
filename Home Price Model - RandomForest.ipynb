```python
#getting the features
features=['Landsize']
X=df[features].values
y=df.Price.values
```

```python
print(X.size, y.size)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y=train_test_split(X,y,random_state=1)
regressor=DecisionTreeRegressor(random_state=0)
regressor.fit(X.reshape(-1,1),y.reshape(-1,1))
```

```python
y_pred=regressor.predict([[165]])
y_pred
```

```python
#plotting the graph
X_grid=np.arange(min(X), max(X), 0.01)
X_grid=X_grid.reshape((len(X_grid), 1))
plt.scatter(X,y,s=20, color='red',label='data' )
plt.plot(X_grid, regressor.predict(X_grid), color='blue', label='max_depth=2')

plt.title('Decision Tree regression')
plt.xlabel('Landsize')
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error
def get_mae(max_leaf_nodes, train_X, test_X, train_y, test_y):
    regressor=DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    regressor.fit(train_X, train_y)
    test_predictions=regressor.predict(test_X)
    test_mae=mean_absolute_error(test_predictions, test_y)
    mae=mean_absolute_error(test_y, test_predictions)
    return(mae)
#write a defloop to compare different tree sizes
for max_leaf_nodes in [5,20,50,250,500,5000]:
    my_mae=get_mae(max_leaf_nodes, train_X,test_X, train_y, test_y)
    print('max leaf nodes: %d \t\t Mean Absolute Error: %d' %(max_leaf_nodes, my_mae))

#fit model using all the data
```
