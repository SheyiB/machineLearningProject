```python
import pandas as pd

file_path = 'C:/Users/E5500/Desktop/melb_data.csv'

home_data = pd.read_csv(file_path)
#from learntools.core import binder
#binder.bind(globals())
#from learntools.machine_learning.ex3 import *

print("Setup Complete")
home_data.describe()
```

```python
home_data = home_data.dropna(axis = 0)
```

```python
y = home_data.Price
```

```python
features_name = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
X = home_data[features_name]
```

```python
X.describe()
```

```python
from sklearn.tree import DecisionTreeRegressor

home_model = DecisionTreeRegressor(random_state = 1)

home_model.fit(X,y)
```

```python
print("making predictions for the first 5 given data")

print(X.head())
predictions = home_model.predict(X.head())
print("  ")
print("The predictions are: ")
print(predictions)
```

```python
home_data.head()
```

```python
#in-sample
from sklearn.metrics import mean_absolute_error

predicted_home_prices = home_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```

```python
#out-sample
from sklearn.model_selection import train_test_split
train_X, val_X, train_y, val_y = train_test_split(X,y, random_state = 0)

home_model = DecisionTreeRegressor()

home_model.fit(train_X, train_y)

val_predictions = home_model.predict(val_X)
print(mean_absolute_error(val_y, val_predictions))
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
```

```python
for max_leaf_nodes in [5, 50,100,300,400,490, 500,5000,50000]:
    my_mae = get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y)
    print("Max leaf nodes: %d  \t\t Mean Absolute Error:  %d" %(max_leaf_nodes, my_mae))
```
